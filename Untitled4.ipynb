```python
dummylist=['cluster1','cluster2','cluster3','cluster4','cluster5','country','genre','rating','month','date']
def get_d(df,l):
    for x in l:
        dummies=pd.get_dummies(df[x],prefix=x,dummy_na=False)
        df=df.drop(x,1)
        df=pd.concat([df,dummies],axis=1)
    return df
df=get_d(df,dummylist)
le=LabelEncoder()
df.score=le.fit_transform(df['score'])
df.name=le.fit_transform(df['name'])
```
